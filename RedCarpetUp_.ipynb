```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
total_customber=1000

#guess that 90% custombers paid bill on time
cust_without_del=95/100

#guess that 2% custombers paid 30days after due date
cust_delay_30days=2/100

#guess that 2% custombers paid 45days after due date
cust_delay_45days=2/100

#guess that 2% custombers paid 60days after due date
cust_delay_60days=2/100

#month average balance
monthly_balance=1000




```

```python
# membership fee is rs20 per card
membership_fees=12*20*total_customber

# rs10 affiliation fee per card 
affiliate_fee=12*10*total_customber

#each card has an interest rate of 15%
interest_rate=15/100

#loss rate is 3%      (3% loss because of custombers not paying credit card bills )
loss_rate=3/100
 
#6.5 cost of funds company must pay to bank
totalcost_to_bank=total_customber*12*1000*0.065/12

#Rs 25 operating cost per card
operating_cost=25*12*total_customber 

```


      Calculation of this given Scenario 

i) The interest rate is simple interest.
ii) Interest is calculated on a daily basis.
iii) The average card balance is the average end of month balance.
iv) All members are more than a year old.

```python
total_days=30   
total_categories=4   #given four groups for diffrent billing date
total_category_type=[]

for customber_category in range(total_categories):
    if customber_category==0:
      total_category_type.append('paid before date')
    else:
      total_category_type.append('paid %d days after due date'%total_days)
      total_days+=15

data=pd.DataFrame({'category':total_category_type},index=np.arange(1,total_categories+1))    
data
# i made a data table of four diffrent categories for billing.
```

```python
total_category_type
```

```python
def interest_calculation(payment_days):
  avg_days_trasaction=15    #average day of transaction as 15th of billing cycle 

  if payment_days==total_category_type[0]:
    interest_per_card=0

  elif payment_days==total_category_type[1]:       #interest for 30 days
    interest_per_card=(avg_days_trasaction+15+30)*monthly_balance*interest_rate/365

  elif payment_days==total_category_type[2]:     #interest for 45 days
    interest_per_card=(avg_days_trasaction+15+45)*monthly_balance*interest_rate/365 

  else:  #interest for 45 days
     interest_per_card=(avg_days_trasaction+15+60)*monthly_balance*interest_rate/365 

  return interest_per_card



def total_int(column):
      payment_days=column[0]
      interest_per_card=column[1]

      if payment_days==total_category_type[0]:
       total_interest=0

 # 
    # Total Interest for all customers who are paying  late by 30 days
      elif payment_days==total_category_type[1]:
         total_interest=12*interest_per_card*cust_delay_30days*total_customber

# Total Interest for all customers who are paying  late by 45 days
      elif payment_days==total_category_type[2]:
         total_interest=12*interest_per_card*cust_delay_45days*total_customber

#Total Interest for all customers who are paying  late by 60 days
      else:
        total_interest=12*interest_per_card*cust_delay_60days*total_customber



      return total_interest
        
```

```python
data['Estimated_Interest_on_Default']=data['category'].apply(interest_calculation).round(2)

data['Total_Interest_1Year']=data[['category','Estimated_Interest_on_Default']].apply(total_int,axis=1)
data['Fixed_Late_Fee_charge']=data['Estimated_Interest_on_Default'].apply(lambda x: 0 if x==0 else 50)
data
```

```python
#calculating total expense

Total_Expense=totalcost_to_bank+operating_cost+(loss_rate*total_customber*monthly_balance)

#calculating total Earnings 

Total_Earnings=membership_fees+affiliate_fee+np.sum(data['Total_Interest_1Year'])+(12*max(data['Fixed_Late_Fee_charge'])*5/100*total_customber)

#calculating total margin

Total_margin=Total_Earnings-Total_Expense

Total_Earnings

```

```python
Total_marginnet_profit_loss=Total_margin/Total_Expense*100
Total_marginnet_profit_loss
```

```python
 def PnL_check(): # Column name based on profit/loss
    if Total_marginnet_profit_loss>=0:
        return 'Profit margin in percentage'
    else:
        return 'Loss margin in percentage'     
```

```python
report=pd.DataFrame({'Total yearly card balance':12*monthly_balance*total_customber,'Earnings':Total_Earnings,'Expenses':Total_Expense,
              'Margin':Total_margin,PnL_check():Total_marginnet_profit_loss.round(2)},index=[' '])

report
```
